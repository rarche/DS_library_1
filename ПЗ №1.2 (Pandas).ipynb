    Задание №1

```python
import pandas as pd
import numpy as np

authors = pd.DataFrame({'author_id': [1, 2, 3],'author_name': ['Тургенев','Чехов','Островский']}, columns = ['author_id', 'author_name'])
book = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3], 'book_title':['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'], 'price':[450, 300, 350, 500, 450, 370, 290]}, columns = ['author_id', 'book_title', 'price'])
```

    Задание №2

```python
authors_price = pd.merge(authors,book, on = 'author_id', how = 'inner')
authors_price
```

    Задание №3

```python
top5 = authors_price.copy()
top5 = top5.sort_values(by = 'price')
top5 = top5.reset_index(drop = 'True')
top5 = top5.tail(5)
top5
```

    Задание №4

```python
authors_max = authors_price.groupby('author_name')['price'].max()
authors_min = authors_price.groupby('author_name')['price'].min()
authors_mean = authors_price.groupby('author_name')['price'].mean()
```

```python
authors_stat_0 = pd.merge(authors_max,authors_min,on='author_name', how='inner')
authors_stat = pd.merge(authors_stat_0,authors_mean,on='author_name', how='inner')
authors_stat.rename(columns = {'price_x':'max_price','price_y':'min_price','price':'mean_price'}, inplace = True)
authors_stat
```

    Задание №5

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
book_info = pd.pivot_table(authors_price, values = 'price', index = 'author_name', columns = 'cover', aggfunc = np.sum, fill_value = 0)
book_info
```
